```python
# Feature engineering functions
def calculate_bmi(weight, height):
    return weight / (height ** 2)

def calculate_physical_activity_index(fcvc, faf, tue):
    return fcvc * faf / (1 + tue)

def calculate_hydration_index(ch2o, tue):
    return ch2o / (1 + tue)

def calculate_age_weight_interaction(age, weight):
    return age * weight

def calculate_pa_hydration_interaction(faf, ch2o, tue):
    return faf * ch2o / (1 + tue)

def calculate_nutrition_physical_activity(fcvc, faf, tue):
    return fcvc * faf / (1 + tue)

def calculate_meal_consumption_index(ncp, fcvc):
    return ncp * fcvc

def calculate_physical_activity_duration(faf, tue):
    return faf * tue

# Feature engineering
num_data['BMI'] = calculate_bmi(num_data['Weight'], num_data['Height'])
num_data['Physical_Activity_Index'] = calculate_physical_activity_index(num_data['FCVC'], num_data['FAF'], num_data['TUE'])
num_data['Hydration_Index'] = calculate_hydration_index(num_data['CH2O'], num_data['TUE'])
num_data['Age_Weight_Interaction'] = calculate_age_weight_interaction(num_data['Age'], num_data['Weight'])
num_data['PA_Hydration_Interaction'] = calculate_pa_hydration_interaction(num_data['FAF'], num_data['CH2O'], num_data['TUE'])
num_data['Nutrition_Physical_Activity'] = calculate_nutrition_physical_activity(num_data['FCVC'], num_data['FAF'], num_data['TUE'])
num_data['Meal_Consumption_Index'] = calculate_meal_consumption_index(num_data['NCP'], num_data['FCVC'])
num_data['Physical_Activity_Duration'] = calculate_physical_activity_duration(num_data['FAF'], num_data['TUE'])

num_data
```

#### Tuning with optuna

```python
import optuna
from sklearn.metrics import make_scorer, accuracy_score

# Define the objective function for Optuna
def objective(trial, classifier, X_train, y_train, X_test, y_test):
    params = {
        'n_estimators': trial.suggest_int('n_estimators', 50, 1000),
        'learning_rate': trial.suggest_loguniform('learning_rate', 0.01, 0.1),
        'max_depth': trial.suggest_int('max_depth', 3, 10),
        'min_samples_split': trial.suggest_int('min_samples_split', 2, 20),
        'min_samples_leaf': trial.suggest_int('min_samples_leaf', 1, 10),
        'random_state': 42, 
        # Add other hyperparameters as needed for each classifier
    }
    
    model = classifier(**params)
    
    # Train the model
    model.fit(X_train, y_train)
    
    # Make predictions on the test set
    y_pred = model.predict(X_test)
    
    # Calculate accuracy
    accuracy = accuracy_score(y_test, y_pred)
    
    return accuracy

# Optimize each classifier
def optimize_classifier(classifier, X_train, y_train, X_test, y_test, n_trials=200):
    study = optuna.create_study(direction='maximize')
    objective_func = lambda trial: objective(trial, classifier, X_train, y_train, X_test, y_test)
    
    study.optimize(objective_func, n_trials=n_trials, show_progress_bar=True)
    
    best_params = study.best_params
    best_model = classifier(**best_params)
    
    # Train the best model on the entire training set
    best_model.fit(X_train, y_train)
    
    return best_model, best_params
```

### Ensemble classifier

```python
from sklearn.ensemble import VotingClassifier

# Initialize your classifiers with random_state=42
gb_classifier = GradientBoostingClassifier(random_state=42, n_estimators=1000)
catboost_classifier = CatBoostClassifier(random_state=42, n_estimators=5000)
lgbm_classifier = LGBMClassifier(random_state=42)
xgboost_classifier = XGBClassifier(random_state=42, n_estimators=2000)

# Create an ensemble using VotingClassifier
ensemble_model = VotingClassifier(
    estimators=[
        ('gb', gb_classifier),
        ('catboost', catboost_classifier),
        ('lgbm', lgbm_classifier),
        ('xgboost', xgboost_classifier)
    ],
    voting='soft'  # Use 'hard' for majority voting or 'soft' for probability averaging
)

# Fit the ensemble model on the training data
ensemble_model.fit(X_train, y_train)

# Make predictions on the test data
y_pred = ensemble_model.predict(X_test)

# Evaluate the ensemble model
accuracy = accuracy_score(y_test, y_pred)
print(f'Ensemble Model Accuracy: {accuracy}')
```
